```python
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt

def plot_val_curve(
    estimator, X_miss, X, title_extra, param_name, param_range, use_log_x, inverse_x, ax
):
    val_curve = validation_curve(
        estimator,
        X=X_miss,
        y=X,
        param_name=param_name,
        param_range=param_range,
        error_score="raise",
    )
    train_scores, test_scores = val_curve

    train_err = 1 - train_scores
    test_err = 1 - test_scores

    mean_train_err = train_err.mean(axis=1)
    std_train_err = train_err.std(axis=1)
    mean_test_err = test_err.mean(axis=1)
    std_test_err = test_err.std(axis=1)

    ax.plot(param_range, mean_train_err, label="train")
    ax.fill_between(
        param_range,
        mean_train_err - std_train_err,
        mean_train_err + std_train_err,
        alpha=0.2,
    )
    
    ax.plot(param_range, mean_test_err, label="test")
    ax.fill_between(
        param_range,
        mean_test_err - std_test_err,
        mean_test_err + std_test_err,
        alpha=0.2,
    )
    
    ax.set_xlabel(param_name)
    ax.set_ylabel("err")
    ax.set_title(f"Validation curve {estimator.__class__.__name__}, {title_extra}")
    if use_log_x:
        ax.set_xscale("log", base=10)
    if inverse_x:
        ax.invert_xaxis()
    ax.legend()

    
def plot_val_curves(models, datas):
    nm, nd = len(models), len(datas)
    fig, ax = plt.subplots(nm, nd)
    
    for i, model in enumerate(models):
        for j, data in enumerate(datas):
            print(f"{(i+1)} / {nm} | {(j+1)} / {nd}")
            
            plot_val_curve(
                model["estimator"],
                data["X_miss"],
                data["X"], 
                data["title_extra"],
                model["param_name"],
                model["param_range"],
                model["use_log_x"],
                model["inverse_x"],
                ax,
            )
    fig.set_size_inches(9 * nd, 6 * nm)
    fig.tight_layout()
    plt.show()
```

```python
models = [
    {
        "estimator": SoftImputer(max_rank=None, verbose=0),
        "param_name": "shrinkage_value",
        "param_range": np.logspace(-3, 3, endpoint=True, num=20),
        "use_log_x": True,
        "inverse_x": False,
    }
]

datas = [
    {
        "X": X,
        "X_miss": X_miss,
        "title_extra": "MCAR %miss=50",
    }
]

plot_val_curves(models, datas)
```
